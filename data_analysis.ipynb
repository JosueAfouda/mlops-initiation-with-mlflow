```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importation des données
data_path = "data/creditcard.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Combien de classes
df.Class.value_counts()
```

```python
# Dataset des transactions normales et Dataset des transactions frauduleuses
anomalies = df[df.Class == 1]
normal = df[df.Class == 0]
print("Anomalies :", anomalies.shape)
print("Normal :", normal.shape)
```

```python
class_counts = df['Class'].value_counts()
class_counts.plot(kind = 'bar', rot=0)
plt.title("Class Distribution")
plt.xticks(range(2), ["Normal", "Anomaly"])
plt.xlabel("Label")
plt.ylabel("Counts")
plt.show()
```

Le graphique montre visuellement l’immense différence dans la fréquence des deux classes.

```python
# Les valeurs de données anormales comprennent-elles des transactions avec des montants excessifs ?
anomalies['Amount'].describe()
```

```python
# Histogramme des montants des transactions frauduleuses
plt.hist(anomalies['Amount'], bins=100, color="red")
plt.title("Histogram of Fraudulent Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

La distribution des montants des transactions anormales est très asymétrique (à droite) et il semble que les transactions anormales comportent des montants peu élevés. En réalité, la plupart des transactions sont inférieures à 105 $ (3ème quartile) ; les transactions frauduleuses ne sont donc pas des transactions de grande valeur.

Qu'en est-il des transactions normales ?

```python
print(normal['Amount'].describe())
# Histogramme des montants des transactions normales
plt.hist(normal['Amount'], bins=100, color="green")
plt.title("Histogram of Normal Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

La distribution des montants des transactions normales est encore plus asymétrique à droite que celle des transactions anormales. La plupart des transactions sont inférieures à 77$ (3è quartile), et certains montants peuvent atteindre des valeurs très élevées, jusqu'à plus de 25000$.

```python
# Montant au fil du temps par classe de transaction
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, sharey=False)
fig.set_size_inches(10, 6)
fig.suptitle('Amount over Time by Transaction Class')
ax1.scatter(anomalies["Time"]/3600, anomalies["Amount"], color="red")
ax1.set_title('Anomaly')
ax2.scatter(normal["Time"]/3600, normal["Amount"], color="green")
ax2.set_title('Normal')
plt.xlabel("Hours elapsed since first transaction")
plt.ylabel("Amount")
plt.show()
```

En analysant ce graphique `Amount over Time by Transaction Class`, voici quelques observations et conclusions sur l’importance de la variable **"Time"** pour différencier les **transactions normales** des **anomalies** :

* Les **transactions normales** (graphe du bas, en vert) sont **denses et continues** dans le temps. On les retrouve **à toutes les heures** depuis le début jusqu'à environ 50 heures.
* Les **transactions anormales** (graphe du haut, en rouge) sont aussi **réparties sur toute la durée**, mais elles semblent **moins nombreuses** et **plus dispersées**.

➡️ **Conclusion** : Il **n'y a pas de période de temps clairement isolée** où seuls des comportements anormaux apparaissent. Les anomalies sont **mélangées dans le temps** avec les transactions normales.

En clair :

> La variable **"Time"** seule ne semble **pas très discriminante** pour identifier si une transaction est anormale ou non.

> Les **anomalies n'apparaissent pas à des moments particuliers** mais plutôt de façon diffuse dans le temps.
